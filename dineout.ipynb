```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#request page source
url = "https://www.dineout.co.in/pune-restaurants/welcome-back"
page = requests.get(url)
page
```

```python
#page source code
page.content
```

```python
soup = BeautifulSoup(page .content, "html.parser")
print(soup.prettify())
```

```python
scraped_name= soup.find_all('a', class_='restnt-name ellipsis')
scraped_name
```

```python
names = []
for name in scraped_name:
    name = name.get_text().replace('\n',"")
    name = name.strip(" ")
    names.append(name)
names
```

```python
scraped_cus= soup.find_all('span', class_='double-line-ellipsis')
scraped_cus
```

```python
cuss = []
for cus in scraped_cus:
    cus = cus.get_text().replace('\n',"").replace('',"").replace('2,000 for 2 (approx)',"").replace('1,600 for 2 (approx)',"").replace('1,100 for 2 (approx)',"").replace('1,200 for 2 (approx)',"").replace('2,300 for 2 (approx)',"").replace('1,500 for 2 (approx)',"").replace('1,700 for 2 (approx)',"").replace('2,400 for 2 (approx)',"").replace('2,900 for 2 (approx)',"").replace('1,300 for 2 (approx)',"").replace('900 for 2 (approx)',"").replace('1,900 for 2 (approx)',"").replace('|',"")
    cus = cus.strip(" ")
    cuss.append(cus)
cuss
```

```python
scraped_loc= soup.find_all('div', class_='restnt-loc ellipsis')
scraped_loc
```

```python
locs = []
for loc in scraped_loc:
    loc = loc.get_text().replace('\n',"")
    loc = loc.strip(" ")
    locs.append(loc)
locs
```

```python
scraped_rat= soup.find_all('div', class_='img-wrap')
scraped_rat
```

```python
rats = []
for rat in scraped_rat:
    rat = rat.get_text().replace('\n',"").replace('Popular for ambience',"").replace('In high demand',"")
    rat = rat.strip(" ")
    
    rats.append(rat)
rats
```

```python
images=[]
for i in soup.find_all("img",class_="no-img"):
    images.append(i['data-src'])
images
```

```python
#scraped data store
data = pd.DataFrame()
data['NAME']= names
data['CUISINE']= cuss
data['ADDRESS']= locs
data['RATING']= rats
data['IMAGE']= images
data.head()
```

```python
data.to_csv('dineout.csv', index=False)
```
