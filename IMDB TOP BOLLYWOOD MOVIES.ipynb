```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#request page source
url = "https://www.imdb.com/india/top-rated-indian-movies/"
page = requests.get(url)
page
```

```python
#page source code
page.content
```

```python
soup = BeautifulSoup(page .content, "html.parser")
print(soup.prettify())
```

```python
scraped_movies= soup.find_all('td', class_='titleColumn')
scraped_movies
```

```python
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n',"")
    movie = movie.strip(" ")
    movies.append(movie)
movies
```

```python
#get rating
scrap_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
scrap_ratings
```

```python
ratings = []
for rating in scrap_ratings:
    rating = rating.get_text().replace('\n', '')
    ratings.append(rating)
ratings
```

```python
#scraped data store
data = pd.DataFrame()
data['Movie Names']= movies
data['Ratings']= ratings
data.head()
```

```python
data.to_csv('IMDB TOP BOLLYWOOD MOVIES.csv', index=False)
```
